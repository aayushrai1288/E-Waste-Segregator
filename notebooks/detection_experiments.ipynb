## 3. Training Analysis
Visualizing the training progress from the `results.csv` generated by YOLOv9.

```python
import pandas as pd
import matplotlib.pyplot as plt

results_path = 'runs/train/exp/results.csv'
if os.path.exists(results_path):
    results = pd.read_csv(results_path)
    plt.figure(figsize=(10, 5))
    plt.plot(results['epoch'], results['mAP_0.5'], label='mAP@0.5')
    plt.plot(results['epoch'], results['mAP_0.5:0.95'], label='mAP@0.5:0.95')
    plt.title('YOLOv9 Training Progress')
    plt.xlabel('Epoch')
    plt.ylabel('mAP')
    plt.legend()
    plt.show()
```
